```python
df = tables[3]
df.columns = ['Name', 'Team(s) played for (years)', 'Pos', 'Pts', 
              'Reb', 'Ast', 'Championships won', 'MVP won', 
              'Finals MVP won', 'All Star', 'Year','Ref']
greatestplayer_df = df[['Name', 'Team(s) played for (years)', 'Pos', 'Pts', 
              'Reb', 'Ast', 'Championships won', 'MVP won', 
              'Finals MVP won', 'All Star', 'Year']]
```

```python
greatestplayer_df = greatestplayer_df.fillna({"Championships won":0})
```
